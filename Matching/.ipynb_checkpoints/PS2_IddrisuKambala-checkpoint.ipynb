```python
##We scale price and population as suggested in the problem: 
radio_merger['ln_price'] = np.log((radio_merger['price']) / 1000)
radio_merger['ln_popu'] = np.log((radio_merger['population_target']) / 1000)
radio_merger = radio_merger.drop(columns = ['population_target', 'price']) ##not needed
radio_merger
```

```python
radio_merger.year.describe() ##we have two years, 2007 and 2008. so we create sub-data for them
```

```python
rd_2007 = radio_merger[radio_merger['year'] == 2007] ##market for 2007
rd_2008 = radio_merger[radio_merger['year'] == 2008] ##market for 2008
years = [rd_2007, rd_2008]
```

```python
##separating buyer and seller characteristics by year
buyer_charact_2008 = rd_2008.loc[:, ['year', 'buyer_lat', 'buyer_long', 'buyer_id', 'corp_owner_buyer', 'num_stations_buyer']]
buyer_charact_2007 = rd_2007.loc[:, ['year', 'buyer_lat', 'buyer_long', 'buyer_id', 'corp_owner_buyer', 'num_stations_buyer']]
target_charact_2008 = rd_2008.loc[:, ['target_lat', 'target_long', 'target_id', 'ln_price', 'hhi_target', 'ln_popu']]
target_charact_2007 = rd_2007.loc[:, ['target_lat', 'target_long', 'target_id', 'ln_price', 'hhi_target', 'ln_popu']]
```

```python
##we have to identify all possible matches or permutations between buyers and targets for years 2007 & 2008!! we import
## intertools for this
import itertools as it
permutations = len(list(it.product(rd_2007.buyer_id.tolist(), rd_2007.target_id.tolist()))) + \
len(list(it.product(rd_2008.buyer_id.tolist(), rd_2008.target_id.tolist())))
```

```python
def comb_2007(buyer_charact_2007, target_charact_2007):
    return(buyer_charact_2007.assign(key=1).merge(target_charact_2007.assign(key=1), on='key').drop('key', 1))
macthes_2007 = comb_2007(buyer_charact_2007, target_charact_2007)
```

```python
def comb_2008(buyer_charact_2008, target_charact_2008):
    return(buyer_charact_2008.assign(key=1).merge(target_charact_2008.assign(key=1), on='key').drop('key', 1))
macthes_2008 = comb_2008(buyer_charact_2008, target_charact_2008)
```

```python
##we now create the counterfactuals
on = ['target_lat', 'target_long', 'ln_price', 'hhi_target', 'ln_popu', 'year', 'buyer_id', 'buyer_lat', \
          'buyer_long', 'num_stations_buyer', 'corp_owner_buyer', 'target_id']
counterfactual_2008 = (macthes_2008.merge(rd_2008[on], on=on, how='left', \
                                          indicator=True).query('_merge == "left_only"').drop('_merge', 1))
on = ['target_lat', 'target_long', 'ln_price', 'hhi_target', 'ln_popu', 'year', 'buyer_id', 'buyer_lat', \
          'buyer_long', 'num_stations_buyer', 'corp_owner_buyer', 'target_id']
counterfactual_2007 = (macthes_2007.merge(rd_2007[on], on=on, how='left', \
                                          indicator=True).query('_merge == "left_only"').drop('_merge', 1))
```

```python
counterfactual = pd.concat([counterfactual_2007, counterfactual_2008])
```

```python
##separating buyer and seller characteristics by year
buyer_charact = ['year', 'buyer_lat', 'buyer_long', 'buyer_id', 'corp_owner_buyer', 'num_stations_buyer']
target_charact = ['target_lat', 'target_long', 'target_id', 'ln_price', 'hhi_target', 'ln_popu']
```

```python
##list comprehension for counterfactuals
countfct_rad = pd.DataFrame()
countfct_rad = [cf[buyer_charact].iloc[b].values.tolist() + cf[target_charact].iloc[t].values.tolist() \
                for cf in years for b in range(len(cf) - 1) for t in range(b + 1, len(cf))]
countfct_rad = pd.DataFrame(countfct_rad, columns = buyer_charact + target_charact)
```

```python
##distance (km) column
countfct_rad['distance'] = countfct_rad.apply(lambda row: geodesic((row.buyer_lat, row.buyer_long), \
                                                                   (row.target_lat, row.target_long)).km, axis = 1)
```

```python
##now we define the payoff functions
def first_payoff(m, i, params):
    '''
   This function helps to calculate the payoffs of the mergers
    Args:
        m: the merger dataset
        i: we use it for iteration purposes 
        params: parameters
    Returns:
        function: payoff
    '''
    alpha = params[1]
    beta = params[1]

    function = m['num_stations_buyer'].iloc[i] * m['ln_popu'].iloc[i] + \
    alpha * m['corp_owner_buyer'].iloc[i] * m['ln_popu'].iloc[i] + beta * m['distance'].iloc[i]

    return(function)
```

```python
# estimating actual payoffs
Params = (0.5, -0.8)
for i in range(len(rd_2007)):
    payoff_actual_2007 = [first_payoff(m = rd_2007, i = i, params = Params)] ##for 2007
    
for i in range(len(rd_2008)):
    payoff_actual_2008 = [first_payoff(m = rd_2008, i = i, params = Params)] ##2008
```

```python
##now counterfactuals
for i in range(0, (len(rd_2007) * (len(rd_2007) - 1))):
    payoff_cf_2007 = [first_payoff(m = countfct_rad, i = i, params = Params)]

for i in range((len(rd_2008) * (len(rd_2008) - 1)), len(countfct_rad)):
    payoff_cf_2008 = [first_payoff(m = countfct_rad, i = i, params = Params)]
```
